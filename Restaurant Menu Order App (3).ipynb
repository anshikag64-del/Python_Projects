```python
n=input("Enter name\n")

menu = {
    'coffee' : 80,
    'salad' : 50,
    'taco' : 30,
    'cold drink' : 40,
    'sandwich' : 20
}

print("Welcome to our cafe!\n")
print("This is our menu for today. Please order and enjoy your dishes\n")
print(menu)

orders = []
total_amount = 0

while True:
    user_input = input("Please select your items from the menu or enter q/quit to exit\n")

    #print(user_input)
    if user_input.lower() in menu.keys():
        total_amount = total_amount + menu[user_input.lower()]
        orders.append(user_input)
        print(f"You have ordered {orders} and the total bill is ${total_amount}\n")
    elif user_input.lower() == 'q' or user_input.lower() == 'quit':
        print("Thank you for your order.\nIt'll be served soon\n")
        tip_bool = input("Do you want to add a tip for the staff? (Yes/No)\n")
        if tip_bool.lower() == 'yes':
            tip_amount = float(input("Enter the tip amount\n"))
            total_amount = total_amount + tip_amount
            print(f"You have ordered {orders} and the total bill is ${total_amount}\n")
        elif tip_bool.lower() == 'no':
            print(f"You have ordered {orders} and the total bill is ${total_amount}\n")
        else:
            print("Enter something valid\n")
        break
    elif user_input.lower() not in menu.keys():
        print("Entered item is not in the menu\n")
    else:
        print("Enter a valid input")
```
